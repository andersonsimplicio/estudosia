```python
import pandas as pd
import pickle
```

```python
with open('database/diabetes_no_stander.pkl','rb') as f:
    x_treinamento,y_treinamento,x_teste,y_teste = pickle.load(f)
```

```python
x_treinamento.shape,y_treinamento.shape
```

```python
df_treinameto = pd.DataFrame(data=x_treinamento,columns=['Pregnancies', 'PlasmaGlucose', 'DiastolicBloodPressure',\
       'TricepsThickness', 'SerumInsulin', 'BMI', 'DiabetesPedigree', 'Age','colesterol','hba1c'])
df_treinameto['Diabetic'] = y_treinamento
```

```python
df_treinameto
```

```python
df_teste = pd.DataFrame(data=x_teste,columns=['Pregnancies', 'PlasmaGlucose', 'DiastolicBloodPressure',\
       'TricepsThickness', 'SerumInsulin', 'BMI', 'DiabetesPedigree', 'Age','colesterol','hba1c'])
df_teste['Diabetic'] = y_teste
```

```python
df_teste
```

```python
df = pd.merge(df_treinameto,df_teste,how='outer')
```
